## 3. Splitting The Data Into X and Y:

```python
# Divide the dataset into X (independent variable) and y (dependent variable)
X = df.iloc[:,:-1].values
y = df.iloc[:,5].values
```

```python
# Implement our classifier based on Linear Regression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)
```

## 4. Train The Model

```python
# Fitting Multiple Linear Regressor to the training set
reg = LinearRegression()
reg.fit(X_train,y_train)
```

## 5. Predicting The Result

```python
# Predicting the Test set results
y_predict = reg.predict(X_test)
```

```python
rss = y_test - y_predict
```

```python
plt.scatter(rss,y_predict)
```

```python
sns.distplot(rss)
```

## 6. Check the Goodness of the Model Using R^2:

```python
# Now we will find the R2 and Adj R2
from sklearn.metrics import r2_score
score = r2_score(y_test, y_predict)
```

```python
score
```

R2 is very very good and very close to 1.. So our prediction is much better...

```python
# Now we'll do plotting
plt.figure(figsize=(15,10))
plt.scatter(y_test,y_predict,color='red')
plt.xlabel("Actual",fontsize=20)
plt.ylabel("Predicted",fontsize=20)
plt.title("Actual vs Predicted")

plt.show()
```

```python
# Now we'll find the difference between the Actual and Predicted Values
predicted_values_df = pd.DataFrame({"Actual Value":y_test,"Predicted Value":y_predict,"Difference":y_test-y_predict})
```

```python
predicted_values_df.tail()
```

## 7. Plotting a graph between Actual and Predicted Values

```python
sns.regplot('Actual Value','Predicted Value',data = predicted_values_df)
```
